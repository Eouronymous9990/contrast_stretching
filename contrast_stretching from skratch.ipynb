```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def contrast_stretching(image):
    # Convert the image to grayscale
    #gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Calculate the minimum and maximum pixel values
    min_val = np.min(image)
    max_val = np.max(image)

    # Perform contrast stretching
    stretched = np.uint8((image - min_val) * (255.0 / (max_val - min_val)))

    # Convert the image back to color (optional)
    #stretched = cv2.cvtColor(stretched, cv2.COLOR_GRAY2BGR)

    return stretched

# Read the input image
image = cv2.imread(r"E:\images.jpeg")

# Apply contrast stretching
result = contrast_stretching(image)

# Display the input and result images
#cv2.imshow('Input', image)
#cv2.imshow('Contrast Stretched', result)
result = contrast_stretching(image)

# Convert BGR to RGB for proper display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
result_rgb = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)

# Display the input and result images
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Input')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(result_rgb)
plt.title('Contrast Stretched')
plt.axis('off')

plt.tight_layout()
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
